```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

def h(x):
   return 100/x

x = np.linspace(-10, 20, 500)

plt.plot(x, h(x), color='black')
plt.xlabel('Elapsed time')                           # label of x axis
plt.ylabel('timeFactor')                             # label of y axis
plt.xticks([0])                                      # x axis range
plt.yticks([0])                                      # y axis range
plt.xlim([0,20])                                     # x axis label numbers
plt.ylim([0,300])                                    # y axis label numbers
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

T = 50
beta = 60

def f(x):
   return -(T/(x + T/beta)) + beta

def g(x):
   return beta - x + x 

x = np.linspace(0 , 25 , 500)

plt.plot(x, f(x), color='black')
plt.plot(x , g(x) , color = 'darkblue' , linestyle='dashed')
plt.xlabel('size')                           # label of x axis
plt.ylabel('sizeFactor')                     # label of y axis
plt.xlim([0,25])                             # x axis range
plt.ylim([0,100])                            # y axis range
plt.xticks([0])                              # x axis label numbers
plt.yticks([0])                              # y axis label numbers
plt.show()
```

```python
from matplotlib import cm

fig = plt.figure()
ax = plt.axes(projection='3d')

xline = np.linspace(0, 6, 30)
yline = np.linspace(0, 6, 30)

xline, yline = np.meshgrid(xline, yline)
def zi(x, y):
    return h(x) * f(y)

zline = zi(xline , yline)

plt.xticks([])                              # x axis label numbers
plt.yticks([])                              # y axis label numbers

ax.contour3D(xline, yline, zline, 200 , rstride=1, cstride=1 , cmap='viridis' , edgecolor='none')                   # more colors can be used using cm.(...) in cmap argument
ax.set_xlabel('timeFactor')
ax.set_ylabel('sizeFactor')
ax.set_zlabel('activity')
```
